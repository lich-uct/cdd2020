```python
from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
import csv
with open('Tomas_Otava_CDG_Epidermal_growth_factor_receptor_erbB1_list_of_compounds.csv','r') as csvfile:
    reader = csv.DictReader(csvfile,delimiter=";")
    rawdata = [x for x in reader]
lig_MOLECULES = [Chem.MolFromSmiles(x['Smiles']) for x in rawdata]
```

```python
suppl = Chem.SDMolSupplier('drugbank.sdf')
drugs = [m for m in suppl if m]
len(drugs)
```

Opět moje oblíbená zcela náhodná molekula 3344

```python
erbB1_ligands_ECFP = [Chem.GetMorganFingerprintAsBitVect(x, 2, nBits=1024)
                     for x in lig_MOLECULES]
drugs_ECFP = [Chem.GetMorganFingerprintAsBitVect(x, 2, nBits=1024) for x in drugs]
```

```python
erbB1_ligands_ECFP_sums = [0]*erbB1_ligands_ECFP[0].GetNumBits()
for key in erbB1_ligands_ECFP:
    for onbit in key.GetOnBits():
        erbB1_ligands_ECFP_sums[onbit] += 1
erbB1_ligands_ECFP_sums
```

```python
drugs_ECFP_sums = [0]*drugs_ECFP[0].GetNumBits()
for key in drugs_ECFP:
    for onbit in key.GetOnBits():
        drugs_ECFP_sums[onbit] += 1
drugs_ECFP_sums
```

```python
erbB1_ligands_ECFP_scaled = [x/len(erbB1_ligands_ECFP) for x in erbB1_ligands_ECFP_sums]
drugs_ECFP_scaled = [x/len(drugs_ECFP) for x in drugs_ECFP_sums]
drugs_ECFP_scaled
```

```python
erbB1_drugbank_differences = [(i, a_b[0] - a_b[1])
                             for i, a_b in enumerate(zip(erbB1_ligands_ECFP_scaled, drugs_ECFP_scaled))]
erbB1_drugbank_differences.sort(key=lambda x: x[1])
erbB1_drugbank_differences[:3], erbB1_drugbank_differences[-3:]
```

Tak nakonec jsem dostal k těmto bitům kde pro DRUGBANK je favorizovaný bit 1 a pro erbB1 set 491.

Tak pojďme zjistit substruktury.

```python
def substructures_of_bit_ecfp4_1024(mol, bit):
    bitinfo = {}
    Chem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024, bitInfo=bitinfo)
    submols = []
    for atom, radius in bitinfo.get(bit, []):
        if radius ==0:
            continue
        hit_atoms = Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atom)
        submols.append(Chem.PathToSubmol(mol, hit_atoms))
    return submols
```

```python
erbB1_substructure_smiles = {}
for x in lig_MOLECULES:
    for submol in substructures_of_bit_ecfp4_1024(x, 491):
        subsmiles = Chem.MolToSmiles(submol)
        erbB1_substructure_smiles[subsmiles] = erbB1_substructure_smiles.get(subsmiles, 0) + 1
erbB1_substructure_smiles
```

```python
drugbank_substructure_smiles = {}
for x in drugs:
    for submol in substructures_of_bit_ecfp4_1024(x, 491):
        subsmiles = Chem.MolToSmiles(submol)
        drugbank_substructure_smiles[subsmiles] = drugbank_substructure_smiles.get(subsmiles, 0) + 1
drugbank_substructure_smiles
```

Pro bit 491 je v setu erbB1 nejčastější substruktura cNc (7943), v drugbanku sice s menším zastoupením (213) je nejčastější tatáž substruktura.

```python
erbB1_substructure_smiles = {}
for x in lig_MOLECULES:
    for submol in substructures_of_bit_ecfp4_1024(x, 1):
        subsmiles = Chem.MolToSmiles(submol)
        erbB1_substructure_smiles[subsmiles] = erbB1_substructure_smiles.get(subsmiles, 0) + 1
erbB1_substructure_smiles
```

```python
drugbank_substructure_smiles = {}
for x in drugs:
    for submol in substructures_of_bit_ecfp4_1024(x, 1):
        subsmiles = Chem.MolToSmiles(submol)
        drugbank_substructure_smiles[subsmiles] = drugbank_substructure_smiles.get(subsmiles, 0) + 1
drugbank_substructure_smiles
```
